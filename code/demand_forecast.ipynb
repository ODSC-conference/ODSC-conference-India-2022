```python
data = pd.read_csv('hist_data.csv')
```

```python
data.head()
```

```python
demand_df = data[1825:]
for i in range(1,16):
    df = data[['date','cust_'+str(i)]]
    df.rename(columns={'cust_'+str(i): 'y', 'date': 'ds'}, inplace=True)
    train = df[:1825]
    valid = df[1825:]
    model = Prophet()
    model.fit(train)
    close_prices = model.make_future_dataframe(periods=len(valid))
    forecast = model.predict(close_prices)
    forecast_valid = forecast['yhat'][1825:]
    #valid['cust_'+str(i)] = forecast_valid.values
    demand_df['cust_'+str(i)] = forecast_valid.values
demand_df = demand_df.drop(columns=['Unnamed: 0','date']).apply(np.ceil)
demand_df=demand_df.transpose()
demand_df.rename(columns={1825:'demand'})
```
